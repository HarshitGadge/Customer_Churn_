```python
data = df.drop(['CustomerID','Name'], axis = 1)
data.columns
```

```python
x = data.iloc[:,0:5]
y = data.iloc[:,-1]
print(x.shape)
print(y.shape)

print(x.columns)
```

```python
x = pd.get_dummies(x)
x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
y_test.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
X_train = mm.fit_transform(X_train)
X_test = mm.transform(X_test)

X_train = pd.DataFrame(X_train)
X_train.head()
```

```python
X_test=pd.DataFrame(X_test)
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuaracy :", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.model_selection import cross_val_score

cvs = cross_val_score(estimator = model, X = X_train, y = y_train, cv = 10)
print(cvs)
```

```python
print("Mean Accuracy :", cvs.mean())
print("Variance :", cvs.std())
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes = (100, 100), activation ='relu',
                      solver = 'adam', max_iter = 50)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

# first hidden layer
model.add(Dense(10, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))

# second hidden layer
model.add(Dense( 8, kernel_initializer = 'uniform', activation = 'relu'))

# third hidden layer
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))

# fourth hidden layer
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))

# fifth hidden layer
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))

# output layer
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the NN
# binary_crossentropy loss function used when a binary output is expected
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model.fit(X_train, y_train, batch_size = 10, epochs=50)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(kernel_init,optimizer,activation,units):
    model = Sequential()
    model.add(Dense(units,kernel_initializer=kernel_init,activation=activation,input_dim=10))
    model.add(Dense(units,kernel_initializer=kernel_init,activation=activation))
    model.add(Dense(1,kernel_initializer=kernel_init,activation='sigmoid'))
    model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
Classifier = KerasClassifier(build_fn=build_classifier)
```

```python
parameters = {'optimizer':['adam','rmsprop','SGD'],
              'activation':['relu','sigmoid'],
              'kernel_init':['uniform','he_uniform','glorot_uniform'],
              'units':[10,12,16,20]}
```

```python
grid_search = GridSearchCV(estimator=Classifier,param_grid=parameters,scoring='accuracy',cv=5)
grid_search = grid_search.fit(X_train,y_train)
```

```python
pred_tuned_ann = grid_search.predict(X_test)
pred_tuned_ann = (pred_tuned_ann>0.7)
```

```python
accuracy_score(y_test,pred_tuned_ann)
```

```python
best_parameters=grid_search.best_params_
best_accuracy = grid_search.best_score_
```
